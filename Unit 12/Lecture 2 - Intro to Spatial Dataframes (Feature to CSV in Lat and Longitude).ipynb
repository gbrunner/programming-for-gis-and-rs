## For this CSV to work with the webapp, the coordinates had to be in Latitude and Longitude. I had to convert that SHAPE JSON into Latitude and Longitude coordinates.

This code effectively strips out the x and y values from the JSON and puts them into new pandas dataframe columns named Lon and Lat.

```python
#x = []
#y = []
for idx,val in enumerate(sdf.iterrows()):
    #print(val[1].SHAPE)
    j = val[1].SHAPE
    #x.append(j['x'])
    #y.append(j['y'])
    sdf.loc[idx,'Lon'] = j['x']
    sdf.loc[idx,'Lat'] = j['y']
    
#sdf['Lon'] = x
#sdf['Lat'] = y
sdf.head()
```

## Now I can save that dataframe to a CSV with a subset of fields

```python
sdf.to_csv(
    path_or_buf=r"stl_crime_wgs_84_10282018.csv", 
    columns=['Crime', 'District', 'Neighborhood', 'ILEADSStreet', 'CADStreet', 'Lon', 'Lat']
)
```
